```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#plt.style.use('classic')
%matplotlib inline

pd.options.display.max_columns = None
```

```python
artdf = pd.read_csv('WikiArtClean.csv')
```

```python
artdf.head()
```

```python
artdfgp = artdf.groupby(by='Year').mean().reset_index()
```

```python
artdfgp
```

```python
positive = artdfgp['Agreeableness'] + artdfgp['Gratitude'] + artdfgp['Happiness'] + artdfgp['Humility'] + artdfgp['Love'] + artdfgp['Optimism'] + artdfgp['Trust']
positivemean = positive.mean()
posroll = positive.rolling(10).mean()


negative = artdfgp['Anger'] + artdfgp['Arrogance'] + artdfgp['Disagreeableness'] + artdfgp['Disgust'] + artdfgp['Fear'] + artdfgp['Pessimism'] + artdfgp['Regret'] + artdfgp['Sadness'] + artdfgp['Shame']
negativemean = negative.mean()
negroll = negative.rolling(10).mean()

neutral = artdfgp['Anticipation'] + artdfgp['Shyness'] + artdfgp['Surprise'] + artdfgp['Neutral']
neutralmean = neutral.mean()
neuroll = neutral.rolling(10).mean()
```

```python
newdata = {'Year':artdfgp['Year'], 'Positive':positive, 'Negative':negative, 'Neutral':neutral, 
          'Positive Emotions':posroll, 'Negative Emotions':negroll, 'Neutral Emotions':neuroll}

rolldata = {'Year':artdfgp['Year'], 'Positive':posroll, 'Negative':negroll, 'Neutral':neuroll}

emotiondf = pd.DataFrame(rolldata)
emotiondf
```

```python
#pd.melt(emotiondf, id_vars=['Year'], value_vars=['Positive Rolling', 'Negative Rolling', 'Neutral Rolling'])
```

```python
fig, ax = plt.subplots(figsize=(16,8))
sns.lineplot(x='Year', y='Average Emotion Value', hue='Emotion Category', data=pd.melt(emotiondf, id_vars=['Year'], 
                                                               value_vars=['Positive',
                                                                           'Negative', 
                                                                           'Neutral'],
                                                              var_name='Emotion Category',
                                                              value_name='Average Emotion Value'))
```
